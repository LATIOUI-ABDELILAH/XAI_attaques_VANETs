## Description du dataset VeReMi

Le dataset contient environ 22 millions d'observations et plusieurs variables décrivant les données du véhicule lui-même (type=2) et les messages reçus d’ autres véhicules (type=3)

Les variables incluent :
- la position des véhicules
- la vitesse
- l'accélération
- la direction
- les valeurs de bruit 

Le dataset contient également deux variables liées aux attaques :

- `attack` : indique si le message correspond à un comportement normal ou malveillant
- `attack_type` : indique le type d'attaque spécifique

## Analyse des attaques dans le dataset

L'analyse des variables liées aux attaques montre que le dataset est relativement équilibré. 
La variable `attack` indique si une observation correspond à un comportement normal ou à une attaque. 
On observe environ 12 millions d'observations normales contre environ 10 millions correspondant à des attaques.

La variable `attack_type` permet d'identifier le type précis de l'attaque. Le dataset contient 19 types d'attaques différents. Le nombre d'observations est assez proche pour chaque type d'attaque.

L'analyse du dataset montre qu'il n'y a pas de valeurs manquantes ni de doublons dans les données
