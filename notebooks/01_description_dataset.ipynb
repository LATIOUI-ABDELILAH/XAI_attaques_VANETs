## Description du dataset VeReMi Extension

Le dataset contient **3 194 808 observations** et **30 colonnes** décrivant les messages BSM (Basic Safety Messages) échangés entre véhicules dans un réseau VANET simulé.

Les variables incluent :
- `sendTime` : le temps d'envoi du message
- `senderPseudo` : l'identifiant pseudonyme du véhicule émetteur
- `messageID` : l'identifiant unique du message
- La position des véhicules (`posx`, `posy`) et leurs valeurs de bruit (`posx_n`, `posy_n`)
- La vitesse (`spdx`, `spdy`) et leurs valeurs de bruit
- L'accélération (`aclx`, `acly`) et leurs valeurs de bruit
- La direction (`hedx`, `hedy`) et leurs valeurs de bruit

La variable cible est :
- `class` : indique le comportement du véhicule — 0 pour un comportement normal, et 1 à 19 pour les différents types d'attaques

## Analyse des attaques dans le dataset

La variable `class` permet d'identifier le type de comportement de chaque véhicule. 
Le dataset contient 20 classes différentes — une classe normale (class=0) et 19 types 
d'attaques (class=1 à 19).

On observe environ 1 900 539 observations normales contre 1 294 269 observations 
correspondant à des attaques.

Le dataset contient 19 types d'attaques différents avec des nombres d'observations 
variables selon le type.

L'analyse du dataset montre qu'il n'y a pas de valeurs manquantes ni de doublons 
dans les données.
